# Demographic Data Visualization
### Population and Geographic Distributions

## Setup & Library Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
```

## Loading the Datasets

```python
# Load population dataset (skipping header lines) and metadata dataset
df_population = pd.read_csv('data/population.csv', skiprows=4)
df_metadata = pd.read_csv('data/metadata.csv')

print(f"Population dataset: {df_population.shape}")
print(f"Metadata dataset: {df_metadata.shape}")
```

## Data Preprocessing & Merging

```python
# Merge on Country Code
df_merged = pd.merge(df_population, df_metadata, on='Country Code', how='inner')

# Filter out regional aggregates (keep only actual countries)
df_countries = df_merged[df_merged['Region'].notna() & (df_merged['Region'] != "")]
print(f"Cleaned countries dataset: {df_countries.shape}")
```

## Visualizing Country Name vs Population (Graph Coordinates)

```python
# Select specific countries to plot on coordinates
target_countries = ['India', 'United States', 'China', 'Brazil', 'Indonesia', 'Japan', 'United Kingdom', 'Germany', 'Canada', 'Australia']
df_selected = df_countries[df_countries['Country Name'].isin(target_countries)].copy()

# Choose year and convert population to millions for readability
year = '2022'
df_selected['Population_M'] = df_selected[year].astype(float) / 1_000_000

# Sort by population to make the coordinates visual and clear
df_selected = df_selected.sort_values(by='Population_M', ascending=False)

# Plot Country Name vs Population as coordinates
plt.figure(figsize=(12, 6))
sns.barplot(
    x='Country Name', 
    y='Population_M', 
    data=df_selected, 
    palette='viridis', 
    hue='Country Name', 
    legend=False
)

plt.title(f'Population of Selected Countries in {year}', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Country Name (X Coordinate)', fontsize=12)
plt.ylabel('Population in Millions (Y Coordinate)', fontsize=12)
plt.xticks(rotation=45)

# Add value labels on top of the bars
for idx, row in enumerate(df_selected.itertuples()):
    plt.text(idx, row.Population_M + 15, f"{row.Population_M:.1f}M", ha='center', fontweight='bold', fontsize=10)

plt.tight_layout()
plt.show()
```

## Categorical Variable Distribution: Countries by Geographic Region

```python
# Compute region counts
region_counts = df_countries['Region'].value_counts()

# Plot distribution
plt.figure(figsize=(12, 6))
sns.barplot(
    x=region_counts.values, 
    y=region_counts.index, 
    palette='plasma', 
    hue=region_counts.index, 
    legend=False
)

plt.title('Distribution of Countries by Geographic Region', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Number of Countries', fontsize=12)
plt.ylabel('Region', fontsize=12)

# Add values on horizontal bars
for idx, val in enumerate(region_counts.values):
    plt.text(val + 0.5, idx, str(val), va='center', fontweight='bold', fontsize=11)

plt.tight_layout()
plt.show()
```
